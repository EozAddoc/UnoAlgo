```python
#visualizer ? 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
path_to_data = '/home/eoz/Desktop/UnoAlgo/ai_vs_human_games.csv'

df = pd.read_csv(path_to_data)
df.head()
```

```python
df.describe()
```

```python
# Difficulty in function of winner
# Convert categorical columns
df["Winner"] = df["Winner"].astype("category")
df["Difficulty"] = df["Difficulty"].astype("category")

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Difficulty", hue="Winner", palette="pastel")
plt.title("Win Distribution by Difficulty")
plt.xlabel("Difficulty Level")
plt.ylabel("Number of Games")
plt.legend(title="Winner")
plt.show()



```
